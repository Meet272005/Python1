```python
def checker(num):
    temp = 0
    while num > 0:
         temp = temp +1
         num = num //10
        
    if temp == 3 :
        return 1
    else :
        return 0
    
    
def palindrome(num):
    temp = num
    
    new = 0
    while num>0 :
        rem = num % 10
        new = (10*new) + rem
       
        num = num //10
        
    if new == temp :
        
        return "palindrome"
    else : 
        return "Not a palindrome"
    
def armstrong(num):
    temp = num
    new =0
    
    while num >0:
        rem = num % 10
        new = new +(rem*rem*rem)
        num = num//10
        
    if new == temp :
        return "Armstrong"
    else :
        return "Not an Armstrong"
    
    
def prime(num):
    temp = 0
    for i in range(1,num+1):
        if num%i == 0:
            temp = temp+1
    
    if temp == 2 :
        return "Prime number"
    else : 
        return "not a prime number"
    
    
def happy(num) : 
    sum = num
    while sum!= 1:
        temp = sum
        sum = 0

        while temp>0 :
            pre = temp % 10
            sum = sum + (pre*pre)
            temp = temp //10

        if sum == 4 :
            break
        
    if sum == 1 :
        return "happy"
    else :
        return "Not happy"


x = int(input("Enter Three digit number : "))

if checker(x) == 1 :
    print("Enter 1 for Palindrome")
    print("Enter 2 for Armstrong")
    print("Enter 3 for Prime")
    print("Enter 4 for Happy number")
    
    y = int(input(" "))
    
    if y == 1 :
        print(palindrome(x))
    elif y == 2:
        print(armstrong(x))
    elif y == 3 :
        print(prime(x))
    elif y == 4 :
        print(happy(x))
    else :
        print("invalide input")
        
else :
    print("Number is not a Three digit number")
```

```python
#wt a program that proms the user to enter the number and stops only when the user enter stop after this print min max and avg
sum= 0
min = 0
max= 0
count = 0
while True : 
    x = input("Enter number : ")
    
    if x == "Stop" : 
        break;
        
    if int(x) >= max : 
        max = int(x)

    if int(x) <= min : 
        min = int(x)
        
    sum = sum + int(x)
    count = count +1
    

print("Sum : ", sum)    
print("Max : ", max)
print("Min : ", min)
print("Avg : ", (sum/count))
```
